```python
import json


def prime_function(f: str = 'is_prime'):
    cache = dict({1: [False, [1, 1]],
                  2: [True,  [1, 2]],
                  3: [True,  [1, 3]],
                  5: [True,  [1, 5]],
                  7: [True,  [1, 7]]})

    def prime(n):
        if not n in cache:
            if not n % 2:
                cache[n] = [False, [2, n // 2]]
            elif not n % 3:
                cache[n] = [False, [3, n // 3]]
            elif not n % 5:
                cache[n] = [False, [5, n // 5]]
            elif not n % 7:
                cache[n] = [False, [7, n // 7]]
            else:
                cache[n] = [True, [1, n]]
            if cache[n][1][-1] != n:
                m = prime(cache[n][1][-1])
                if not m[0]:
                    cache[n][1] += m[1]
                    cache[n][1] = sorted(cache[n][1])
        return cache[n]

    def is_prime(n):
        return prime(n)[0]

    if f == 'is_prime':
        return is_prime
    elif f == 'prime':
        return prime
    else:
        return None


is_prime = prime_function('is_prime')
prime = prime_function('prime')

for n in range(9999, 1, -1):
    print(f"{n}: {prime(n)}")
```
